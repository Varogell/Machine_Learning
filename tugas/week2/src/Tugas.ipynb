```python
import pandas as pd

data = 'wbc.csv'
wbc = pd.read_csv(data)
wbc.head
```

```python


# Menentukan variabel yang tidak dapat digunakan
kolom = ['id', 'Unnamed: 32']

# Menghapus variabel yang tidak dapat digunakan 
df = wbc.drop(columns=kolom)

# Menampilkan dataframe yang telah dipisahkan
df.head()
 
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Mengambil kolom numerik
kolom_numerik = df.select_dtypes(include=['float64', 'int64']).columns

# Membuat objek StandardScaler
scaler = StandardScaler()

# Melakukan standarisasi pada kolom numerik
df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])

# Menampilkan dataframe setelah standarisasi
df.head()
```

```python
# Split data
from sklearn.model_selection import train_test_split

# Rasio yang akan kita gunakan adalah 80:20
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])
# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.2, random_state=0, stratify=df_unseen['diagnosis'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
print(f'Jumlah label data val:\n{df_val.diagnosis.value_counts()}')
```
